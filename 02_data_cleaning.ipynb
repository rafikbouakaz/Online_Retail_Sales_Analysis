#  Data Cleaning

In this notebook, we clean the online retail dataset by handling missing values, removing invalid transactions, and preparing the data for analysis.

## Import Libraries

In this section, we import the necessary Python libraries for data manipulation and analysis.

```python
import pandas as pd
import numpy as np

# Display settings
pd.set_option('display.max_columns', None)

print(" Libraries imported successfully")
```

##  Load Dataset

We load the online retail dataset and display the first few rows to understand its structure.

```python
# Load dataset
df = pd.read_csv('data/online_retail_II.csv', encoding='ISO-8859-1')

print("Dataset loaded successfully")
df.head()
```

## Initial Data Exploration

We examine the dataset size, structure, and data types to understand its characteristics.

```python
# Dataset shape and info
print(f"Shape: {df.shape[0]:,} rows Ã {df.shape[1]} columns\n")
df.info()
```

## Check Missing Values

We identify missing values in the dataset to determine which columns require cleaning.

```python
# Check missing values
missing = df.isnull().sum()

print("Missing values per column:")
print(missing)
```

##  Handle Missing Customer IDs

Rows with missing Customer ID are removed since they cannot be used for customer-based analysis.

```python
# Remove rows with missing Customer ID
before = len(df)

df = df.dropna(subset=['Customer ID'])

after = len(df)

print(f"Removed {before - after} rows with missing Customer ID")
print(f"New shape: {df.shape}")
```

##  Remove Duplicate Records

Duplicate rows are removed to ensure data accuracy and avoid double counting.

```python
# Remove duplicate rows
before = len(df)

df = df.drop_duplicates()

after = len(df)

print(f"Removed {before - after} duplicate rows")
```

##  Remove Cancelled Transactions

Cancelled orders are identified by invoices starting with 'C' and are removed since they do not represent actual sales.

```python
# Remove cancelled orders (Invoice starting with 'C')
before = len(df)

df = df[~df['Invoice'].astype(str).str.startswith('C')]

after = len(df)

print(f"Removed {before - after} cancelled transactions")
print(f"New shape: {df.shape}")
```

##  Remove Invalid Values

We remove rows with negative or zero values in Quantity and Price, as they do not represent valid transactions.

```python
df = df[(df['Quantity'] > 0) & (df['Price'] > 0)]
```

##  Convert Date Format

The InvoiceDate column is converted to datetime format for time-based analysis.

```python
# Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

print(" Converted InvoiceDate to datetime")
```

##  Feature Engineering

We create a new column called TotalPrice to represent the total revenue for each transaction.

```python
# Create TotalPrice column
df['TotalPrice'] = df['Quantity'] * df['Price']

print("Created TotalPrice column")
df.head()
```

## Final Data Overview

We examine the cleaned dataset to verify the results of the cleaning process.

```python
# Final check
print(f"Final dataset shape: {df.shape}")
df.describe()
```

```python
# Save cleaned dataset
df.to_csv('data/online_retail_cleaned.csv', index=False)

print(" Cleaned dataset saved successfully")
```

##  Cleaning Summary

- Removed rows with missing CustomerID
- Removed duplicate transactions
- Removed cancelled orders
- Removed invalid values (negative or zero)
- Converted InvoiceDate to datetime
- Created TotalPrice feature

The dataset is now clean and ready for exploratory data analysis.
